```python
# 将向量视为标量值组成的列表
x = torch.arange(4)
x
```

```python
# 通过张量的索引来访问任一元素
x[3]
```

```python
# 访问张量的长度
len(x)
```

```python
# 只有一个轴的张量，形状只有一个元素
x.shape
```

```python
# 通过指定两个分量m和n创建一个形状为mxn的矩阵
A = torch.arange(20).reshape(5, 4)
A
```

```python
# 矩阵的转置
A.T
```

```python
# 对称矩阵（symmetric matrix）A=A.T
B = torch.tensor([[1,2,3], [2,0,4], [3,4,5]])
B
```

```python
# 向量是标量的推广，矩阵是向量的推广，构建具有多轴的数据结构
X = torch.arange(24).reshape(2,3,4)
```

```python
B == B.T
```

```python
# 给定具有相同形状的任何两个张量，任何按元素二元运算的结果都将是相同形状的张量
A = torch.arange(20, dtype=torch.float32).reshape(5,4)
B = A.clone()
A, A+B
```

```python
# 两个矩阵按元素乘法称为hadamard product
A * B
```

```python
a = 2
X = torch.arange(24).reshape(2,3,4)
a + X, (a * X).shape
```

```python
A = torch.arange(20*2).reshape(2, 5, 4)
A.shape, A.sum(), A
```

```python
# 指定求和汇总张量的轴
A_sum_axis0 = A.sum(axis=0)
A_sum_axis0, A_sum_axis0.shape
```

```python
print(A.sum(axis=[0, 2]))
A
```

```python
# 计算总和活均值时保持轴数不变
sum_A = A.sum(axis=1, keepdims=True)
sum_A.shape
```

```python
# 通过广播将A除以sum_A
A / sum_A
```
