```python
# 给定具有相同形状的任何两个张量，任何按元素二元运算的结果都将是相同形状的张量
A = torch.arange(20, dtype=torch.float32).reshape(5,4)
B = A.clone()
A, A+B
```

```python
# 两个矩阵按元素乘法称为hadamard product
A * B
```

```python
A = torch.arange(20*2).reshape(2, 5, 4)
A.shape, A.sum(), A
```

```python
# 指定求和汇总张量的轴
A_sum_axis0 = A.sum(axis=0)
A_sum_axis0, A_sum_axis0.shape
```

```python
print(A.sum(axis=[0, 2]))
A
```

```python
# 计算总和活均值时保持轴数不变
sum_A = A.sum(axis=1, keepdims=True)
sum_A.shape
```

```python
# 通过广播将A除以sum_A
A / sum_A
A
```

```python
# 某个轴计算A元素的累积总和
A.cumsum(axis=0)
```

```python
# 点积是相同位置的按元素乘积的和
y = torch.ones(4, dtype=torch.float32)
x = torch.arange(4, dtype=torch.float32)
x, y, torch.dot(x, y)
```

```python
# 矩阵向量Ax是一个长度为m的列向量，其ith元素是点积aiTx, mv -- matrix-vector
A = torch.arange(20).reshape(5, 4).float()
A.shape, x.shape, A, x, torch.mv(A, x)
```

```python
# mm -- matrix-matrix
B = torch.ones(4, 3)
torch.mm(A, B), B.dtype
```

```python
# L2范数是向量元素平方和的平方根
u = torch.tensor([3.0, -4.0])
torch.norm(u)
```

```python
# 矩阵的弗罗贝尼乌斯 Frobenius norm 是矩阵元素的平方和的平方根
torch.norm(torch.ones((4, 9)))
```
